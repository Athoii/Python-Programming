**Write a simple program that has the following functionalities.**

> Write a function called addition() that will take two integer parameters as input. Add the two numbers and return the result.

> Write a function called multiplication() that will take two integer parameters as input. Multiply the two numbers and return the result.

> Write a function called takeInput() that takes no parameter. This function does the follwoings -

>> a. It will take two numbers as input from the user and return the values.

>> b. It will take the choise of operation (e.g., addition or multiplication) as input form the user and return the choice.

> Now call the takeInput() function to take two numbers and choise of operation as input.

> now based on the choise  call the appropriate function to perform the operation and get the result.

> Finally, print the two input values and the result.

```python
def addition(num1, num2):
  sum = num1 + num2
  return sum

def multiplication(num1,num2):
  multi = num1*num2
  return multi

def takeInput():
  n1 = int(input('Number 1: '))
  n2 = int(input('Number 2: '))

  choice = input('Choice(A or M): ')

  return [n1,n2,choice]


inputs = takeInput()


if inputs[2] == 'A':
  result = addition(inputs[0],inputs[1])
  print(f'{inputs[0]} + {inputs[1]} = {result}')

elif inputs[2] == 'M':
  result = multiplication(inputs[0],inputs[1])
  print(f'{inputs[0]} X {inputs[1]} = {result}')

'''
Easier Approach:
x,y,z = takeInput()


if z == 'A':
  result = addition(x,y)
  print(f'{x} + {y} = {result}')

elif z == 'M':
  result = multiplication(x,y)
  print(f'{x} X {y} = {result}')
'''
```
